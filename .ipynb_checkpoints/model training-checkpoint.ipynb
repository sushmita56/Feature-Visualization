```python
DIRECTORY  = r'/content/dogscats/train'
Categories = ['dogs', 'cats']
```

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
for category in Categories:
    folder = os.path.join(DIRECTORY, category)
    label = Categories.index(category)
    for img in os.listdir(folder):
        img_path = os.path.join(folder, img)
        img_arr = cv2.imread(img_path)
        plt.imshow(img_arr)
        plt.xlabel(category)
        break
    break
```

```python
print(img_arr.shape)
```

```python
img_size = 70

newim_arr = cv2.resize(img_arr, (img_size, img_size))
plt.imshow(newim_arr)
plt.show()
```

```python
training_data = []

def get_training_data():
    for category in Categories:
        folder = os.path.join(DIRECTORY, category)
        label = Categories.index(category)
        for img in os.listdir(folder):
            img_path = os.path.join(folder, img)
            img_arr = cv2.imread(img_path)
            newim_arr = cv2.resize(img_arr, (img_size, img_size))
            training_data.append([newim_arr, label])

get_training_data()
```

```python
print(len(training_data))
```

```python
import random

random.shuffle(training_data)
```

```python
X = []
y = []

for features, labels in training_data:
    X.append(features)
    y.append(label)
```

```python
X = np.array(X)
y = np.array(y)
```

```python
X.shape, y.shape
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
```

```python
cnn = Sequential([
                  tf.keras.layers.Conv2D(64, kernel_size=(3,3), activation = 'relu', input_shape=(70,70,3)),
                  tf.keras.layers.MaxPooling2D((2,2)),

                  tf.keras.layers.Conv2D(64, kernel_size=(3,3), activation = 'relu'),
                  tf.keras.layers.MaxPooling2D((2,2)),

                  tf.keras.layers.Flatten(),
                  tf.keras.layers.Dense(50, activation='relu'),
                  tf.keras.layers.Dense(1, activation='sigmoid')
])

cnn.compile(loss = 'binary_crossentropy',
            optimizer = 'adam',
            metrics = ['accuracy'])

cnn.fit(X, y, epochs = 2, validation_split = 0.1, batch_size=32)
```
