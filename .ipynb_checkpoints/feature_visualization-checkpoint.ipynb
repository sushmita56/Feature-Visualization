```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
model = tf.keras.models.load_model('/content/drive/MyDrive/DL/Cat-Dog/trained_model')
```

```python
model.summary()
```

```python
import os
import cv2
import matplotlib.pyplot as plt
cv_img = []
img_size = 70
folder = r'/content/drive/MyDrive/DL/Cat-Dog/testdata'
for file in os.listdir(folder):
    n= cv2.imread(os.path.join(folder,file)) 
    newim_arr = cv2.resize(n, (img_size, img_size))
    newim_arr = cv2.cvtColor(newim_arr, cv2.COLOR_RGB2GRAY)
    cv_img.append(newim_arr)
    break
```

```python
img = np.array(np.squeeze(cv_img))
```

```python
plt.imshow(img)
```

```python
img.shape
```

```python
img = np.reshape(img, (1, 70,70,1))
```

```python
for layer in model.layers:
  if 'conv' not in layer.name:
    continue
  filters, biases = layer.get_weights()
  print(layer.name, filters.shape)
```

```python
filters, biases = model.layers[0].get_weights()

fig = plt.figure(figsize = (16,8))
columns = 8
rows = 4
n_filters = rows * columns
for i in range(1, n_filters + 1):
  f = filters[:, :, :,i-1]
  fig = plt.subplot(rows, columns, i)
  fig.set_xticks([])
  fig.set_yticks([])
  plt.title('filter: ' + str(i))
  plt.imshow(f[:, :, 0], cmap='ocean')
plt.show()
```

```python
# Visualizing first convolutional layer
conv2d_2_output =  tf.keras.models.Model(inputs = model.input, outputs = model.get_layer('conv2d_2').output)
conv2d_2_features = conv2d_2_output.predict(img)


fig = plt.figure(figsize=(16,8))
# plt.title('Convolutional Layer 1")
columns = 8
rows = 4
for i in range(columns * rows):
  fig.add_subplot(rows, columns, i+1)
  plt.axis('Off')
  plt.title('conv_output: ' + str(i))
  plt.imshow(conv2d_2_features[0, :, :, i])
plt.show()
```

```python
# Visualizing first maxpooling layer
max_pooling2d_2_output =  tf.keras.models.Model(inputs = model.input, outputs = model.get_layer('max_pooling2d_2').output)
max_pooling2d_2_features = max_pooling2d_2_output.predict(img)


fig = plt.figure(figsize=(16,8))
columns = 8
rows = 4
for i in range(columns * rows):
  fig.add_subplot(rows, columns, i+1)
  plt.axis('Off')
  plt.title('Pooling: ' + str(i))
  plt.imshow(max_pooling2d_2_features[0, :, :, i])
plt.show()
```

```python
filters, biases = model.layers[2].get_weights()

fig = plt.figure(figsize = (16,8))
columns = 8
rows = 4
n_filters = rows * columns
for i in range(1, n_filters + 1):
  f = filters[:, :, :,i-1]
  fig = plt.subplot(rows, columns, i)
  fig.set_xticks([])
  fig.set_yticks([])
  plt.title('filter: ' + str(i))
  plt.imshow(f[:, :, 0], cmap='ocean')
plt.show()
```
